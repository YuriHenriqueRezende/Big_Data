```python
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt


leucocitos = np.array([2000, 4000, 5000, 6500])
plaquetas = np.array([100000, 20000, 80000, 145000])
linfocitos = np.array([2.3, 4.5, 6.5, 4.4])


X = np.column_stack((leucocitos, plaquetas, linfocitos))


contaminacao = 0.25  
iso_forest = IsolationForest(contamination=contaminacao)
iso_forest.fit(X)


predicoes = iso_forest.predict(X)

plt.figure(figsize=(8, 6))

plt.scatter(X[predicoes == 1][:, 0], X[predicoes == 1][:, 1], c='blue', label='Observações normais')

plt.scatter(X[predicoes == -1][:, 0], X[predicoes == -1][:, 1], c='red', label='Anomalias')

plt.xlabel('Leucócitos')
plt.ylabel('Plaquetas')
plt.title('Detecção de Anomalias com Isolation Forest')
plt.legend()
plt.show()
```
