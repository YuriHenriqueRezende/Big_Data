```python
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt


corrente_eletrica = np.array([5, 10, 14, 2, 1.5, 6])
tempo = np.array([1, 2, 4, 6, 7, 10])


X = np.column_stack((corrente_eletrica, tempo))


contaminacao = 0.2  
iso_forest = IsolationForest(contamination=contaminacao)
iso_forest.fit(X)


predicoes = iso_forest.predict(X)


plt.figure(figsize=(8, 6))


plt.scatter(X[predicoes == 1][:, 1], X[predicoes == 1][:, 0], c='blue', label='Observações normais')

plt.scatter(X[predicoes == -1][:, 1], X[predicoes == -1][:, 0], c='red', label='Anomalias')

plt.xlabel('Tempo')
plt.ylabel('Corrente Elétrica')
plt.title('Detecção de Anomalias com Isolation Forest')
plt.legend()
plt.show()
```
