```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn import tree
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
notas = np.array([8,6,5,9,4,3,7,2,1,5,6,9,8])
classes = np.array(['bom','ruim','ruim','bom','pessimo','pessimo',
 'bom','pessimo', 'pessimo','ruim','ruim','bom','bom'])
```

```python
notas_treino,notas_teste,classes_treino,classes_teste = train_test_split(notas.reshape(-1,1),classes.reshape(-1,1),test_size=0.2,random_state=42)
```

```python
modelo = DecisionTreeClassifier()
```

```python
modelo.fit(notas_treino,classes_treino)
```

```python
fig = plt.figure(figsize=(10,8))
tree.plot_tree(modelo,feature_names=notas.tolist(),class_names=classes.tolist(),filled=True)
```

```python
sns.boxplot(x=classes,y=notas)
plt.title('Notas e classificação boletim')
plt.xlabel('Classificação')
plt.ylabel('Notas')
```

```python
sns.barplot(x=classes,y=notas)
plt.title("notas e classificação das bebidas")
```
