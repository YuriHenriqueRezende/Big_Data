```python
x, y = iris.data, iris.target
classfier = tree.DecisionTreeClassifier()
classfier.fit(x,y)
fig = plt.figure(figsize=(30,25))
tree.plot_tree(classfier, feature_names=iris.feature_names, class_names=iris.target_names.tolist(), filled=True)
plt.savefig('dt.png')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(iris.data[:,0],iris.data[:,2],c=iris.target)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[2])
plt.tight_layout()
plt.show()
```
