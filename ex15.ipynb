```python
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt


leucocitos = np.array([2000, 4000, 5000, 6500])
plaquetas = np.array([100000, 20000, 80000, 145000])
linfocitos = np.array([2.3, 4.5, 6.5, 4.4])

X = np.column_stack((leucocitos, plaquetas, linfocitos))

eps = 1000 
min_samples = 2  
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
labels = dbscan.fit_predict(X)

grupos = set(labels)
n_grupos = len(grupos)

plt.figure(figsize=(8, 6))
for grupo in grupos:
    pontos_grupo = X[labels == grupo]
    plt.scatter(pontos_grupo[:, 0], pontos_grupo[:, 1], label=f'Grupo {grupo}')

plt.xlabel('Leucócitos')
plt.ylabel('Plaquetas')
plt.title('Grupos de Pacientes Identificados pelo DBSCAN')
plt.legend()
plt.show()

for grupo in grupos:
    print(f'Grupo {grupo}: Pacientes = {len(labels[labels == grupo])}')
```
